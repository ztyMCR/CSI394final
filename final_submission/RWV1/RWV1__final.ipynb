```python
import copy

n = 0
m = 0

i = 1
total = 0
zeroL = []
zeroOff = []
firL = []
secL = []
firOff = []
secOff = []
tempR = 0
tempG = 0
tempB = 0

with open('colors.ppm', 'r') as fin:
  
  first_line = fin.readline()
  second_line = fin.readline()
  third_line = fin.readline()
  print ("The first line is "+ first_line)
  print ("The second line is " + second_line)
  print ("The third line is "+ third_line)
  width,height = second_line.split()
  print ("width is", width ," and height is", height)

  for line in fin:
    zeroL.append(int(line))
    
#done reading

width = int(width)
height = int(height)
finalwidth = width
finalheight = height
rwvN = input("Please give the N for subtraction")
print("You entered: ", rwvN)
Ntimes  = int(rwvN)

total = Ntimes

print("the new zeroL has ",len(zeroL)," elements")

while(Ntimes>0):
    
#average width and adding offset
    
    for num in zeroL:
        tempColor = int(num)
       
        if i == 1:
            tempR = tempColor
            i = i+1
            
        elif i == 2:
            tempG = tempColor
            i = i+1
           
        elif i ==3:
            tempB = tempColor
            i = i+1
        elif i == 4:
            avgr = (tempR+tempColor)/2
            firL.append(avgr)
            firOff.append(tempR-avgr)
            i = i+1
        elif i == 5:
            avgg = (tempG+tempColor)/2
            firL.append(avgg)
            firOff.append(tempG-avgg)
            i = i+1
        elif i == 6:
            avgb = (tempB+tempColor)/2
            firL.append(avgb)
            firOff.append(tempB-avgb)
            i = 1
        n = n+1
        
        if n%(int(finalwidth)*3) == 0:
            #print("the len of firL is ",len(firL))
            firL.extend(firOff)
            #print("the len of firL is ",len(firL))
            #print("the len of firOff is ",len(firOff))
            del firOff[:]
            #print("merge!!! ",m," at n = ",n)
            
            m = m+1
     
    #print("there are ",n," numbers")
    
    #print("firstL has ",len(firL))
    #print("width half picture done")

    #average height and adding offsets
    
    

    for i in range(int(finalheight/2)):
        for j in range(finalwidth):
            firstr2 = i*finalwidth*3*2 +3*j
            secondr2 = i*finalwidth*3*2 +3*j+3*finalwidth
            firstg2 = firstr2+1
            secondg2 = secondr2+1
            firstb2 = firstg2+1
            secondb2 = secondg2+1
            avgr2 = (firL[firstr2]+firL[secondr2])/2
            avgg2 = (firL[firstg2]+firL[secondg2])/2
            avgb2 = (firL[firstb2]+firL[secondb2])/2
            secL.append(avgr2)
            secL.append(avgg2)
            secL.append(avgb2)
#week3
            secOff.append(firL[firstr2]-avgr2)
            secOff.append(firL[firstg2]-avgg2)
            secOff.append(firL[firstb2]-avgb2)
            
        
    secL.extend(secOff)
    #print("secOff has ", len(secOff))
    #print("secL has ",len(secL))
       
        
            
    
#reset vars
    del zeroL[:]
    zeroL = copy.deepcopy(secL)
    print("the new zeroL has ",len(zeroL)," elements")
    print("the secL has ",len(secL)," elements")
    del firL[:]
    del secL[:]
    Ntimes = Ntimes -1

    del zeroOff[:]
    del firOff[:]
    del secOff[:]

    i = 1
    print("the new width ,height is",width," ",height)
    
    print("one loop done")
    


        
with open('rwv1output.rwv1', 'w') as fout:
    fout.write(first_line)
    fout.write(str(finalwidth))
    fout.write(" ")
    fout.write(str(finalheight)+"\n")
    
    fout.write(third_line)
    for i in range(len(zeroL)):
        fout.write(str(int(zeroL[i]))+"\n")

        
print("done writing")
```
